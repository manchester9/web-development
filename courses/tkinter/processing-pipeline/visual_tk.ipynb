```python
#####################
# IMPORTING LIBRARIES
#####################
import tkinter as tk
from tkinter import *
from PIL import Image, ImageTk
```

```python
###################################
# ADDING OBJECTS TO THE MAIN WINDOW
###################################
class Window(Frame):
    def __init__(self, master = None):
        Frame.__init__(self, master)
        
        self.master = master
        self.init_window()
      
       
    def init_window(self):
        self.master.title('Portfolio Optimizer')
        self.pack(fill = BOTH, expand = 1)
        
        submitButton = Button(self, text = 'Submit')
        submitButton.place(x = 350, y = 650)
        
        investorProfileButton = Button(self, text = 'Risk Profile')
        investorProfileButton.place(x = 20, y = 50)
        
        stockPick01Button = Button(self, text = 'Stock Pick 01')
        stockPick01Button.place(x = 20, y = 100)
                
        stockPick02Button = Button(self, text = 'Stock Pick 02')
        stockPick02Button.place(x = 20, y = 150)
                
        stockPick03Button = Button(self, text = 'Stock Pick 03')
        stockPick03Button.place(x = 20, y = 200)
                
        stockPick04Button = Button(self, text = 'Stock Pick 04')
        stockPick04Button.place(x = 20, y = 250)
                
        stockPick05Button = Button(self, text = 'Stock Pick 05')
        stockPick05Button.place(x = 20, y = 300)
        
#         quitButton = Button(self, text = 'Quit' ,command = self.client_exit)
#         quitButton.place(x = 0, y = 0)
        
        menu = Menu(self.master) # Menu of the master
        self.master.config (menu = menu) # Instance of the menu
 
        file = Menu(menu)
        file.add_command(label = 'Exit', command = self.client_exit)
        menu.add_cascade(label = 'File', menu = file)
        
        edit = Menu(menu)
        edit.add_command(label = 'Undo')
        menu.add_cascade(label = 'Undo', menu = edit)
        
        run = Menu(menu)
        run.add_command(label = 'Run optimization')
        menu.add_cascade(label = 'Run', menu = run)
        
        windows = Menu(menu)
        windows.add_command(label = 'Windows')
        menu.add_cascade(label = 'Windows', menu = windows)
        
        version = Menu(menu)
        version.add_command(label = 'Version')
        menu.add_cascade(label = 'Version', menu = version)
        
        image = Menu(menu)
        image.add_command(label = 'Show Image', command=self.showImg)
        image.add_command(label = 'Show Text', command=self.showTxt)
        menu.add_cascade(label = 'Image', menu = image)
        
        
        label = Label(self.master)
        self.master.config(label = label)
        
        label.pack('test')
        
    def showImg(self):
        load = Image.open('pic.png')
        render = ImageTk.PhotoImage(load)
        
        img = Label(self, image = render)
        img.image = render
        img.place(x = 0, y = 0)
   
    def showTxt(self):
        text = Label(self, text = 'Hey there')
        text.pack()
        

#         self.label = Label(master, text = 'Portfolio Optimizer')
#         self.label.pack()

        
#         master = Tk()
#         variable = StringVar(master)
#         variable.set('Risk Averse Investor')
        
#         w = OptionMenu(master, variable, 'Risk Averse Investor', 'Risky Investor')
#         w.pack()
                
    def client_exit(self):
        exit()       
```

```python
###########################################
# lAUNCHING TKINTER MAIN WINDOW AND OBJECTS 
###########################################
root = Tk()
root.geometry("700x700")
app = Window(root)
root.mainloop()
```
