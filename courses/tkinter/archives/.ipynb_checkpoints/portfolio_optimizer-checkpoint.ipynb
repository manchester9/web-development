```python
from tkinter import *
import random 
import time
```

```python
######################
# CREATING MAIN WINDOW
######################
window = Tk() # create a Tk() object
# Window = Window(window) # create a window object by passing app into a Window class
window.geometry("1600x800") # size of the window width
window.title('Portfolio Optimizer') # title of the window

text_Input = StringVar()

################
# FRAME OBJECT
################
Tops = Frame(window, width = 1800, height = 200, relief = SUNKEN)
Tops.place(x = 0, y = 0) # bg = 'powder blue'

f1 = Frame(window, width = 900, height = 600, relief = SUNKEN)
f1.place(x = 0, y = 200) 

f2 = Frame(window, width = 900, height = 600, relief = SUNKEN)
f2.place(x = 900, y = 200 )
# f2.pack(side = RIGHT)

############
# FUNCTIONS
############
# Button functions
def btnClick(numbers):
    global operator
    operator = operator + str(numbers)
    text_Input.set(operator)

# Time info
local_time = time.asctime(time.localtime(time.time()))
    
##############
# LABEL OBJECT
##############
label_info = Label(Tops, font = ('arial', 20, 'bold'), text = 'PORTFOLIO OPTIMIZER', fg = 'Steel Blue', bd = 10, anchor = 'w')
label_info.place(x = 600, y = 0)

label_info = Label(Tops, font = ('arial', 10, 'bold'), text = local_time, fg = 'Steel Blue', bd = 10, anchor = 'w')
label_info.place(x = 600, y = 100)

#############
# TEXT OBJECT
#############
# Text display
# txt_display = Entry(f2, font = ('arial', 20, 'bold'), textvariable = text_Input, bd = 30, insertwidth = 4,
#                     bg = 'powder blue', justify = 'right')
# txt_display.grid(columnspan = 4)

################
# BUTTON OBJECT
################
btn7 = Button(f1, padx = 10, pady = 10, fg = 'black', font = ('arial', 15, 'bold'), 
              text = 'Investor Profile', bg = 'powder blue', command = lambda: btnClick(7)).place(x = 10, y = 150)

btn8 = Button(f1, padx = 10, pady = 10, fg = 'black', font = ('arial', 15, 'bold'), 
              text = 'Return Need', bg = 'powder blue', command = lambda: btnClick(7)).place(x = 10, y = 250)

btn9 = Button(f1, padx = 10, pady = 10, fg = 'black', font = ('arial', 15, 'bold'), 
              text = 'Stock pick 01', bg = 'powder blue', command = lambda: btnClick(7)).place(x = 10, y = 350)

# Addition = Button(f2, padx = 16, pady = 16, fg = 'black', font = ('arial', 20, 'bold'), 
#               text = '7', bg = 'powder blue', command = lambda: btnClick('+')).grid(row = 2, column = 3)

#################
# DROPDOWN OBJECT
#################










```
